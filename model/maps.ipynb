# Concentration maps study

```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl


# Build a discrete colorbar
cmap = mpl.colors.ListedColormap(["#7fcdbb", "#1d91c0", "#0c2c84"])
cmap.set_under('#eeeeee')
norm = mpl.colors.BoundaryNorm([2, 20, 100, 200], cmap.N) 
```

## Select maps to show

```python
titles = ['0.3 µL/min, 20 min',
          '3 µL/min, 20 min',
          '0.3 µL/min, 60 min',
          '3 µL/min, 60 min',
          '10 µg, 0.3 µL/min, 204 min',
          '10 µg, 3 µL/min, 109 min']

files = ['../2022-09-05 0.3 uLpmin/c_20min.npy',
         '../2022-09-03 3 uLpmin/c_20min.npy',
         '../2022-09-05 0.3 uLpmin/c_60min.npy',
         '../2022-09-03 3 uLpmin/c_60min.npy',
         '../2022-09-05 0.3 uLpmin/c_10ug.npy',
         '../2022-09-03 3 uLpmin/c_10ug.npy']

mask_files = ['../2022-09-05 0.3 uLpmin/mask_rot.npy',
         '../2022-09-03 3 uLpmin/mask_rot.npy',
         '../2022-09-05 0.3 uLpmin/mask_rot.npy',
         '../2022-09-03 3 uLpmin/mask_rot.npy',
         '../2022-09-05 0.3 uLpmin/mask_rot.npy',
         '../2022-09-03 3 uLpmin/mask_rot.npy']
```

## Load files

```python
cs = []
masks = []
for file in files:
    cs.append(np.load(file))
for file in mask_files:
    masks.append(np.load(file))

for i, c in enumerate(cs):
    c[np.isnan(masks[i])] = np.nan
```

```python
ncols = len(titles)

plt.rcParams["figure.figsize"] = (3*ncols,6)
figure, ax = plt.subplots(ncols=ncols)

#con = ax[0].contourf(c_20s[0], [2, 20, 100, 200], origin='image', cmap='YlGnBu')
ax[0].set_title(titles[0])
ax[0].axis('equal')
ax[0].axis('off')
con =ax[0].imshow(cs[0], cmap=cmap, norm=norm)
figure.colorbar(con, ax=ax[0], label="Concentration range (µg/mL)")
for i, title in enumerate(titles[1:]):
    ax[i+1].imshow(cs[i+1], cmap=cmap, norm=norm)
    ax[i+1].set_title(titles[i+1])
    ax[i+1].axis('equal')
    ax[i+1].axis('off')
plt.show()
```
